```python
# Import the choice function of the random module
# https://stackoverflow.com/questions/306400/how-to-randomly-select-an-item-from-a-list
import random
# Assign to a list the 3 possible options: 'stone', 'paper' or 'scissors'.
options=["stone", "paper", "scissors"]

# Assign a variable to the maximum number of games: 1, 3, 5, etc ...
games = 3
games_played = 0

# Assign a variable to the number of games a player must win to win.
# Preferably the value will be based on the number of maximum games
points_for_win = 2


# Define a function that randomly returns one of the 3 options.
# This will correspond to the play of the machine. Totally random.
def machine(options):
    return options[random.randrange(0, 3)];

# Define a function that asks your choice: 'stone', 'paper' or 'scissors'
# you should only allow one of the 3 options. This is defensive programming.
# If it is not stone, paper or scissors keep asking until it is.
def player(options):
    
    option_p = input("Write an option: 'stone', 'paper' or 'scissors'")
    
    while option_p not in options:
        option_p = input("Write an option: 'stone', 'paper' or 'scissors'")
        
    return(option_p)

# Define a function that resolves a combat.
# Returns 0 if there is a tie, 1 if the machine wins, 2 if the human player wins
def result(machine, player):
    if (machine == player):
        return 0
    elif((machine == "stone" and player == "scissors") or (machine == "scissors" and player == "paper") or (machine == "paper" and player == "stone")):
        return 1
    else:
        return 2

# Define a function that shows the choice of each player and the state of the game
# This function should be used every time accumulated points are updated

def choice(machine, player, wins_player, wins_machine, games_played, games):
    print("Machine selection:", machine)
    print("Player selection:", player)
    print("Wins player: ", wins_player)
    print("Wins machine: ", wins_machine)
    print("Games:", games_played, "of", games)

# Create two variables that accumulate the wins of each participant

wins_machine = 0
wins_player = 0

# Create a loop that iterates while no player reaches the minimum of wins
# necessary to win. Inside the loop solves the play of the
# machine and ask the player's. Compare them and update the value of the variables
# that accumulate the wins of each participant.

while (wins_machine != points_for_win) and (wins_player != points_for_win) and ( games_played != games):
    
    select_machine = machine(options)
    select_player = player(options)
    
    if(result(select_machine,select_player) == 1):
        wins_machine+=1
        games_played += 1
    elif(result(select_machine,select_player) == 2):
        wins_player+=1
        games_played += 1
    elif(result(select_machine,select_player) == 0): 
        games_played += 1
        
    choice(select_machine, select_player, wins_player, wins_machine, games_played, games)
    
# Print by console the winner of the game based on who has more accumulated wins
if wins_machine>wins_player:
    print("The PC wins")
elif wins_player>wins_machine:
    print("You win")
elif wins_machine==wins_player:
    print("Tie")
```

```python
# Import the choice function of the random module
import random

# Define a function that asks for an odd number on the keyboard, until it is not valid
# will keep asking
def number():
    n=int(input("Write a number of games: "))
    while n%2==0:
        n=int(input("Write a number of games: "))
    return n


# Assign a list of 5 possible options.

options = ["stone", "paper", "scissors", "lizard", "spock"]

# Assign a variable to the maximum number of games: 1, 3, 5, etc ...
# This time the previously defined function is used

games = number()


# Assign a variable to the number of games a player must win to win.
# Preferably the value will be based on the number of maximum games
games_to_win=(games/2)+0.5

# Define a function that randomly returns one of the 5 options.
# This will correspond to the play of the machine. Totally random.
def machine(options):
    return options[random.randrange(0, 5)];

# Define a function that asks your choice between 5
# you should only allow one of the 5 options. This is defensive programming.
# If it is not valid, keep asking until it is valid.
def player(options):
    option_p = input("Write an option: ")
    
    while option_p not in options:
        option_p = input("Write an option: ")
        
    return(option_p)

# Define a function that resolves a combat.
# Returns 0 if there is a tie, 1 if the machine wins, 2 if the human player wins
# Now there are more options

def result(machine, player):
    if (machine == player):
        return 0
    elif (machine == "stone" and player == "scissors") or (machine == "scissors" and player == "paper") or (machine == "paper" and player == "stone") or (machine=="stone" and player=="lizard") or (machine=="lizard" and player=="spock") or (machine=="lizard" and player=="paper") or (machine=="spock" and player=="stone") or (machine=="spock" and player=="scissors") or (machine=="scissors" and player=="lizard") or (machine=="paper" and player=="spock"):
        return 1
    else:
        return 2

    
# Define a function that shows the choice of each player and the state of the game
# This function should be used every time accumulated points are updated


def choice (machine, player, wins_machine, wins_player, games_played, games):
    print("Machine selection:", machine)
    print("Player selection:", player)
    print("Machine wins:", wins_machine)
    print("Player wins:", wins_player)
    print("Games played:", games_played)
    print("Total games:", games)

    
# Create two variables that accumulate the wins of each participant

wins_player=0
wins_machine=0
    

# Create a loop that iterates while no player reaches the minimum of wins
# necessary to win. Inside the loop solves the play of the
# machine and ask the player's. Compare them and update the value of the variables
# that accumulate the wins of each participant.
games_played=0
while wins_player<games_to_win and wins_machine<games_to_win and games_played<games:
    machine_selection=machine(options)
    player_selection=player(options)
    
    if result(machine_selection, player_selection)==0:
        games_played+=1
    elif result(machine_selection, player_selection)==1:
        games_played+=1
        wins_machine+=1
    elif result(machine_selection, player_selection)==2:
        games_played+=1
        wins_player+=1
    choice(machine_selection, player_selection, wins_machine, wins_player, games_played, games)

# Print by console the winner of the game based on who has more accumulated wins

if wins_machine>wins_player:
    print("The machine wins")
elif wins_machine<wins_player:
    print("You win")
elif wins_machine==wins_player:
    print("It's a tie")
```
